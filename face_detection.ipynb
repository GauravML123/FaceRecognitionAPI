```python
#face recognition

import requests


url = "https://face-verification2.p.rapidapi.com/FaceVerification"

files1 = {
	"photo1": open('photos/ts1.jpg', 'rb'),
	"photo2": open('photos/ts2.jpg', 'rb')
}

files2 = {
	"photo1": open('photos/ts1.jpg', 'rb'),
	"photo2": open('photos/mc1.jpg', 'rb')
}


headers = {
	"X-RapidAPI-Key": "YOUR API KEY FROM RAPID AI",
	"X-RapidAPI-Host": "face-verification2.p.rapidapi.com"
}

response1 = requests.post(url, files=files1, headers=headers)
response2 = requests.post(url, files=files2, headers=headers)

response1 = response1.json()
message1 = response1['data']['resultMessage']
score1 = response1['data']['similarPercent']

response2 = response2.json()
message2 = response2['data']['resultMessage']
score2 = response2['data']['similarPercent']
```

```python
import matplotlib.pyplot as plt
img1 = plt.imread("photos/ts1.jpg")
img2 = plt.imread("photos/ts2.jpg")

fig = plt.figure()

fig.add_subplot(1,2,1)
plt.imshow(img2)
plt.axis(False)
fig.add_subplot(1,2,2)
plt.imshow(img1)
plt.axis(False)
plt.title(f"similairty score: {score1:2f}")
```
